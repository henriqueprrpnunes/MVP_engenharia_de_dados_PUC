```python
import pandas as pd
import glob
import os

# Pasta onde estão os arquivos
folder = r"C:\Users\henri\OneDrive\Documentos\Base dados MVP Engenharia de dados\Contratos_Públicos"

# Lista todos os arquivos .csv da pasta
all_files = glob.glob(os.path.join(folder, "*.csv"))
print(f"{len(all_files)} arquivos encontrados.")

dfs = []

for file in all_files:
    print(f"Lendo: {file}")

    # Tenta ler com UTF-8, se der erro de codificação tenta Latin1
    try:
        df = pd.read_csv(file, sep=";", encoding="utf-8")
    except UnicodeDecodeError:
        df = pd.read_csv(file, sep=";", encoding="latin1")
    except Exception as e:
        print(f"Erro inesperado no arquivo {file}: {e}")
        continue  # pula o arquivo com erro

    # Adiciona nome do arquivo de origem
    df["arquivo_origem"] = os.path.basename(file)

    dfs.append(df)

# Concatena tudo
if dfs:  
    df_final = pd.concat(dfs, ignore_index=True)
else:
    raise ValueError("Nenhum arquivo CSV válido foi carregado.")

print("Concatenação concluída. Exibindo primeiras linhas:")
display(df_final.head())

# Caminho de saída
caminho_saida = r"C:\Users\henri\OneDrive\Documentos\Base dados MVP Engenharia de dados\contratos_publicos_unificado.csv"

# Salva CSV final
df_final.to_csv(caminho_saida, index=False, encoding="utf-8")
```
